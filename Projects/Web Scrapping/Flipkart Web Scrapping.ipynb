# Data Visualization

```python
plt.bar(d['rating'],d['price'])
plt.title('price Vs rating',fontsize=20,color='red')
plt.ylabel('price',fontsize=15)
plt.xlabel('rating',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.bar( d['product name'], d['rating'],color='green')
plt.xlabel('product name',fontsize=15)
plt.title('product name Vs rating',fontsize=20,color='blue')
plt.ylabel('rating',fontsize=15)
plt.show()
```

```python
print(d['product name'].value_counts())
plt.figure(figsize=(15,4))
sns.countplot(x=d['product name'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(df['product name'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=df['product name'])
plt.xticks(rotation=90)
plt.show()
```
