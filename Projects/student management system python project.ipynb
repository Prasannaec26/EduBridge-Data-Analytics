```python
global studentlists
studentlists = ['reema','roopavathy','arpit','sanober','prasanna']


         
def studentslist():
    print('       \nwelcome to Student Management System')
    print("""

    Enter 1 : To view sorted Student's List 
    Enter 2 : To Add New Student name
    Enter 3 : To Search Student name
    Enter 4 : To Remove Student name
    Enter 5 : To find Avg & Grade of any Student
    Enter 6 : To find Avg & Grade of all Students

        """)
    

    
    userInput=(int(input('enter the number: ')))
    if(userInput == 1):
        print("\nSorted Students List\n")
        studentlists.sort()
        for students in studentlists:
            print("=> {}".format(students))
        
        
    elif(userInput == 2):
        newstd=input('Enter the student: ')
        if(newstd in studentlists):
            print('\nThe student name {} is already exist\n'.format(newstd))
        else:
            studentlists.append(newstd)
            print('\nThe student name {} is added successfully\n'.format(newstd))
            for students in studentlists:
                print("=> {}".format(students))
        
        
    elif(userInput == 3):
        searchstd=input('Enter the student: ')
        if(searchstd in studentlists):
            print('\nThe student name {} is found\n'.format(searchstd))
        else:
            print('\nThe student name {} is not found\n'.format(searchstd))
        
        
    elif(userInput == 4):
        removestd=input('Enter the student: ')
        if(removestd in studentlists):
            studentlists.remove(removestd)
            print('\nThe student name {} is removed from the list\n'.format(removestd))
        else:
             print('\nThe student name {} is not found in the list\n'.format(removestd))
                      
                
    elif(userInput == 5):
        searchstd=input('Enter the student: ')
        print('\nMarks of {} is: \n'.format(searchstd))
        if(searchstd in studentlists):
            tot=0
            for i in range(5):
                num=int(input('Enter the mark: '))
                tot=tot+num
                avg=tot/5
            print('\nAverage mark of {} is: '.format(searchstd),avg)
            if(avg>90):
                print('Grade of {} is A'.format(searchstd))
            elif(avg>80):
                print('Grade of {} is B'.format(searchstd))
            elif(avg>65):
                print('Grade of {} is C'.format(searchstd))
            elif(avg>40):
                print('Grade of {} is D'.format(searchstd))
            else:
                print('Grade of {} is F'.format(searchstd))
        else:
            print('student name is not in the studentlists')
            
            
    elif(userInput == 6):
        print('\nAverage mark\n')
        for students in studentlists:
            print('\nMarks of {} is: \n'.format(students))
            tot=0
            for i in range(5):
                num=int(input('Enter the mark: '))
                tot=tot+num
                avg=tot/5
            print('\nAverage mark of {} is: '.format(students),avg)
            if(avg>90):
                print('Grade of {} is A'.format(students))
            elif(avg>80):
                print('Grade of {} is B'.format(students))
            elif(avg>65):
                print('Grade of {} is C'.format(students))
            elif(avg>40):
                print('Grade of {} is D'.format(students))
            else:
                print('Grade of {} is F'.format(students))
                
                       
    else:
        print('Enter valid number:?')
        

            
studentslist()


def runAgain():
    run= input("\nwant To Run Again Y/n: ")
    if(run.lower()== 'y' or run.upper()== 'Y'):
        studentslist()
        runAgain()
    elif(run.lower()== 'n' or run.upper()== 'N'):
        print('\nbye')
    else:
        print('Type "Y" or "N" only')
    

runAgain()
```
