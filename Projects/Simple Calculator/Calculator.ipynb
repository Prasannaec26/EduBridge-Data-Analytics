## Calculator using tkinter

```python
from tkinter import *

window=Tk()
window.title("Simple Calculator")

e=Entry(window,justify = RIGHT,width=48,borderwidth=10)
e.place(x=530,y=50)

l=Label(window,text="Calculator",fg='white',bg='grey',font=('Times New Roman',20),width=20)
l.place(x=532,y=0)

def Click(number):
    new_number = e.get() + str(number)
    e.delete(0, END)
    e.insert(0, new_number)
    
def Clear():
    e.delete(0, END)    
    
def Del():
    new_number = e.get()
    lenght = len(new_number)-1
    e.delete(lenght, END)

def Dot(number):
    new_number = e.get() + str(".") 
    e.delete(0, END)
    e.insert(0, new_number)
      
def Add(first_number):
    global f_num
    global math
    math = "addition"
    f_num = first_number
    e.delete(0, END)
    
def Subtract(first_number):
    global f_num
    global math
    math = "subtraction"
    f_num = first_number
    e.delete(0, END)

def Multiply(first_number):
    global f_num
    global math
    math = "multiplication"
    f_num = first_number
    e.delete(0, END)

def Divide(first_number):
    global f_num
    global math
    math = "division"
    f_num = first_number
    e.delete(0, END)
    
def Mod(first_number):
    global f_num
    global math
    math = "Mod"
    f_num = first_number
    e.delete(0, END)
    
def Equal(second_number):
    num_1 = f_num
    if math == "addition":
        e.delete(0, END)
        e.insert(0, float(num_1) + float(second_number))
    if math == "subtraction":
        e.delete(0, END)
        e.insert(0, float(num_1) - float(second_number))
    if math == "multiplication":
        e.delete(0, END)
        e.insert(0, float(num_1) * float(second_number))
    if math == "division":
        e.delete(0, END)
        e.insert(0, float(num_1) / float(second_number))
    if math == "Mod":
        e.delete(0, END)
        e.insert(0, float(num_1) % float(second_number))    
    
       
buttonclear=Button(window,text="C",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=Clear)
buttondel=Button(window,text="DEL",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=Del)
buttondot=Button(window,text=".",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Dot(e.get()))
buttonmod=Button(window,text="%",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Mod(e.get()))
buttondivide=Button(window,text="/",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Divide(e.get()))
button7=Button(window,text="7",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(7))
button8=Button(window,text="8",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(8))
button9=Button(window,text="9",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(9))
buttonmultiply=Button(window,text="X",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Multiply(e.get()))
button4=Button(window,text="4",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(4))
button5=Button(window,text="5",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(5))
button6=Button(window,text="6",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(6))
buttonsubtract=Button(window,text="-",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Subtract(e.get()))
button1=Button(window,text="1",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(1))
button2=Button(window,text="2",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(2))
button3=Button(window,text="3",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Click(3))
buttonadd=Button(window,text="+",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Add(e.get()))
button0=Button(window,text="0",fg='black',bg='white',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=17, command=lambda: Click(0))
buttonequal=Button(window,text="=",fg='white',bg='grey',font=('Times New Roman',10),activebackground='yellow',activeforeground='purple',width=6, command=lambda: Equal(e.get()))


buttonclear.place(x=540,y=120)
buttonmod.place(x=620,y=120)
buttondel.place(x=700,y=120)
buttondivide.place(x=780,y=120)
button7.place(x=540,y=160)
button8.place(x=620,y=160)
button9.place(x=700,y=160)
buttonmultiply.place(x=780,y=160)
button4.place(x=540,y=200)
button5.place(x=620,y=200)
button6.place(x=700,y=200)
buttonsubtract.place(x=780,y=200)
button1.place(x=540,y=240)
button2.place(x=620,y=240)
button3.place(x=700,y=240)
buttonadd.place(x=780,y=240)
button0.place(x=540,y=280)
buttondot.place(x=700,y=280)
buttonequal.place(x=780,y=280)


window.mainloop()
```
