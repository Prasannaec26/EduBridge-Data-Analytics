```python
d.drop(['CustomerID','Genre','Age'],axis=1,inplace=True)
d
```

```python
d.rename({'Annual Income (k$)':'Annual_Income','Spending Score (1-100)':'Spending_Score'},axis=1,inplace=True)
```

```python
plt.scatter(data=d,x='Annual Income',y='Spending Score',color='green')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=5)
```

```python
y=model.fit_predict(d)
```

```python
y
```

```python
d['cluster'] = y
```

```python
d
```

```python
s_0=d[d.cluster==0]
s_0
```

```python
s_1=d[d.cluster==1]
s_1
```

```python
s_2=d[d.cluster==2]
s_2
```

```python
s_3=d[d.cluster==3]
s_3
```

```python
s_4=d[d.cluster==4]
s_4
```

```python
sse=[]
kr=range(1,10) 
for i in kr:
    kn=KMeans(n_clusters=i)
    kn.fit(d)
    sse.append(kn.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('sum of square error')
plt.plot(kr,sse)
plt.show
```

```python
plt.scatter(data=s_0,x='Annual_Income',y='Spending_Score',color='black')
plt.scatter(data=s_1,x='Annual_Income',y='Spending_Score',color='green')
plt.scatter(data=s_2,x='Annual_Income',y='Spending_Score',color='blue')
plt.scatter(data=s_3,x='Annual_Income',y='Spending_Score',color='yellow')
plt.scatter(data=s_4,x='Annual_Income',y='Spending_Score',color='pink')
plt.show()
```

```python
plt.scatter(data=s_0,x='Annual_Income',y='Spending_Score',color='red')
plt.scatter(data=s_1,x='Annual_Income',y='Spending_Score',color='green')
plt.scatter(data=s_2,x='Annual_Income',y='Spending_Score',color='blue')
plt.scatter(data=s_3,x='Annual_Income',y='Spending_Score',color='yellow')
plt.scatter(data=s_4,x='Annual_Income',y='Spending_Score',color='pink')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=300,color='black',marker='*',label='centroid')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('customer segmentation')
plt.show()
```
