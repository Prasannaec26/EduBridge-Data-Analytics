```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
```

```python
df=datasets.load_boston()
df
```

```python
df.data
```

```python
df.feature_names
```

```python
df.target
```

```python

data = pd.DataFrame({'CRIM':df.data[:,0],
                    'ZN':df.data[:,1],
                     'INDUS':df.data[:,2],
                    'CHAS':df.data[:,3],
                   'NOX':df.data[:,4],
                   'RM':df.data[:,5],
                   'AGE':df.data[:,6],
                   'DIS':df.data[:,7],
                   'RAD':df.data[:,8],
                   'TAX':df.data[:,9],
                   'PTRATIO':df.data[:,10],
                  'B':df.data[:,11],
                    'LSTAT':df.target})
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
x=data.drop(['LSTAT'],axis=1)
x
```

```python
y=data.LSTAT
y
```

```python
from sklearn.preprocessing import scale
```

```python
x=scale(x)
x
```

```python
y=scale(y)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(random_state=3)
```

```python
model.fit(x_train,y_train)
```

```python
r_score=model.score(x_train,y_train) 
```

```python
r_score
```

```python
r_score1=model.score(x_test,y_test)
r_score1
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_predict)
```

```python
mse
```

```python
x_ax = range(len(y_test))
```

```python
x_ax
```

```python
plt.plot(x_ax,y_test,label='original')
plt.plot(x_ax,y_predict,label='Predicted')
plt.title('Actual Vs Predicted')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='best',fancybox=True,shadow=True)
plt.show()
```
