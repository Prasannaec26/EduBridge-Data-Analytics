```python
x=d.drop(['salary'],axis=1)
x
```

```python
y=d.salary
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.metrics import *
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import *
import xgboost as xgb
import time
```

```python
def model_create(model, x_train, y_train, x_test, y_test):
    i_time=time.time()
    model.fit(x_train,y_train)
    model.score(x_test,y_test)
    y_predict=model.predict(x_test)
    acc_s=accuracy_score(y_test,y_predict)
    t=time.time()-i_time
    print('Time Taken: {}'.format(t))
    print('accuracy score: {}'.format(acc_s))
    print(classification_report(y_test,y_predict))
    plot_confusion_matrix(model, x_test, y_test)
    return model, acc_s,t
```

```python
model_LR=LogisticRegression()
model_LR, accuracy_LR,t= model_create(model_LR, x_train, y_train, x_test, y_test)
```

```python
model_RF=RandomForestClassifier()
model_RF, accuracy_RF,t= model_create(model_RF, x_train, y_train, x_test, y_test)
```

```python
model_DC=DecisionTreeClassifier()
model_DC, accuracy_DC,t= model_create(model_DC, x_train, y_train, x_test, y_test)
```

```python
model_KN=KNeighborsClassifier()
model_KN, accuracy_KN,t= model_create(model_KN, x_train, y_train, x_test, y_test)
```

```python
model_AB=AdaBoostClassifier(n_estimators=50,learning_rate=1)
model_AB,accuracy_AB,t_AB= model_create(model_AB, x_train, y_train, x_test, y_test)
```

```python
model_ABC=AdaBoostClassifier(base_estimator=RandomForestClassifier(n_estimators=100),n_estimators=50,learning_rate=1)
model_ABC,accuracy_ABC,t_ABC= model_create(model_ABC, x_train, y_train, x_test, y_test)
```

```python
model_BC= BaggingClassifier(base_estimator=RandomForestClassifier(),n_estimators=100,max_samples=0.8,oob_score=True,random_state=1)
model_BC,accuracy_BC,t_BC= model_create(model_BC, x_train, y_train, x_test, y_test)
```

```python
model_XG=xgb.XGBClassifier()
model_XG,accuracy_XG,t_XG= model_create(model_XG, x_train, y_train, x_test, y_test)
```
