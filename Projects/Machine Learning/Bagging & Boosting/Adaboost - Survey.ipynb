**Machine Learning**

```python
d['GENDER'] = d['GENDER'].replace({'M':0,'F':1})
```

```python
d['LUNG_CANCER'] = d['LUNG_CANCER'].replace({'YES':1,'NO':0})
```

```python
def age(agevalue):
    if agevalue < 30:
        return 1
    elif agevalue <50:
        return 2
    else:
        return 3
```

```python
d['AGE'] = d['AGE'].apply(age)
d
```

```python
x=d.drop(['LUNG_CANCER'],axis=1)
x
```

```python
y=d.LUNG_CANCER
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

## DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

## RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

## AdaBoost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier 
```

```python
model_ab=AdaBoostClassifier(n_estimators=50,learning_rate=1)
```

```python
model_ab.fit(x_train,y_train)
```

```python
model_ab.score(x_test,y_test)
```

```python
y_predict=model_ab.predict(x_test)
y_predict
```

## Accuracy score

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```

## confusion matrix & Report

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(model_ab,x_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```
