```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
d['GENDER'].unique()
```

```python
d['GENDER'] = d['GENDER'].replace({'M':0,'F':1})
```

```python
d['LUNG_CANCER'] = d['LUNG_CANCER'].replace({'YES':1,'NO':0})
```

```python
def age(agevalue):
    if agevalue < 30:
        return 1
    elif agevalue <50:
        return 2
    else:
        return 3
```

```python
d['AGE'] = d['AGE'].apply(age)
d
```

```python
x=d.drop(['LUNG_CANCER'],axis=1)
x
```

```python
y=d.LUNG_CANCER
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

## Cross_val_score

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_cross = cross_val_score(DecisionTreeClassifier(),x_train,y_train,cv=5)
```

```python
model_cross
```

```python
model_cross.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bg = BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=100,max_samples=0.8,oob_score=True,random_state=1)
```

```python
bg.fit(x_train,y_train)
```

```python
bg.oob_score_
```
