## Exploring Data

### To drop columns

```r
d2=select(d,-sales,-num_of_ratings)
head(d2)
```

## To rename columns

```r
d3=rename(d2,color=base_color,screensize=screen_size,displaysize=display_size,rearcamera=num_rear_camera,frontcamera=num_front_camera,batterycapacity=battery_capacity,price=sales_price,discount=discount_percent)
head(d3)
```

## Filter particular data

```r
d4=filter(d3,model=='iPhone 12')
d4
dim(d4)
```

```r
d5=d3[d3$discount>0.10,c(1:14)]
d5[order(d5$discount, decreasing = TRUE), ]
dim(d5)
```

```r
brand=d3[d3$brand=='Samsung',]
brand
dim(brand)
```

## Summarize data

```r
d6=d3 %>% group_by(model) %>% summarize(Total_price=sum(price)) %>% arrange(desc(Total_price))
head(d6)
dim(d6)
```

```r
d7=d3 %>% group_by(brand) %>% summarize(Total_price=sum(price)) %>% arrange(desc(Total_price))
d7
```

```r
disprice=summarize(d3,dis_price=price-(discount*price)) 
head(disprice)
dim(disprice)
```

## To add new columns

```r
d3['disprice']<-disprice
head(d3)
```

```r
d8=d3[d3$brand=='Apple' & d3$model=='iPhone 12 Mini',]
d8
dim(d8)
```
